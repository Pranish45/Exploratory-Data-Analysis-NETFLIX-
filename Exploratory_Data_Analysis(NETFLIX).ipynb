```python
df.fillna({'Director' : 'Director NOT specified','Cast' : 'Missing', 'Country' : 'Missing', 'Date_Added': 'Missing', 'Rating' : 'Missing', 'Duration': 'Missing' },inplace= True)
```

```python
ax = df['Release_Year'].value_counts() \
        .head(10) \
        .plot(kind = 'barh', title= 'Top 10 Release Years',)
ax.set_xlabel('Count')
ax.set_ylabel('Release Year')
plt.show()
```

```python
df['Country'].value_counts().head()
```

```python
df.groupby('Release_Year')['Type'].value_counts()
```

```python
Lg = df.groupby('Release_Year')['Type'].value_counts() \
    .unstack() \
    .plot(kind = 'line', stacked = True, figsize=(15,6))
Lg.set_xlabel('Release Year')
Lg.set_ylabel('Count')
Lg.set_title('Rise of Shows over Decades')
plt.show()
```

# **Step 4: Ask a Question about the data**
---

Q1. Show TOP 5 Countries with Highest number of TV show & Movies on NETFLIX?

```python
bx = df['Country'].value_counts().head(6) \
    .to_frame() \
    .query('Country != "Missing"') \
    .sort_values(by = 'Country') \
    .plot(kind = 'bar', title = 'Top 5 Countries')
bx.set_xlabel('Country')
bx.set_ylabel('Count')
plt.show()
display(pd.DataFrame(df['Country'].value_counts().head(6)))
```

Q2. What is the total number of Movies on NETFLIX from INDIA?

```python
df.query('Country == "India" & Type == "Movie"').head()
```

```python
df.query('Country == "India" & Type == "Movie"').shape[0]
```

Q3. Show TOP 5 Directors who gave highest number of shows to NETFLIX?

```python
df['Director'].value_counts().head(6)
```

```python
df.query('Director != "Director NOT specified"').value_counts() \
.groupby('Director') \
.agg(['count']) \
.query('count >= 14') \
.sort_values(('count'), ascending = False)
```

```python
df.query('Director != "Director NOT specified"').value_counts() \
.groupby('Director') \
.agg(['count']) \
.query('count >= 14') \
.sort_values(('count'), ascending = True) \
.plot(kind = 'barh', figsize = (10,6))
plt.show()
```
